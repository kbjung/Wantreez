```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import platform
from matplotlib import font_manager, rc

# 한글 폰트 지정
path = "C:/Windows/Fonts/malgun.ttf"
if platform.system() == "Windows":
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
elif platform.system()=="Darwin":
    rc('font', family='AppleGothic')
else:
    print("Unknown System")
```

```python
path = "./"
file_list = os.listdir(path)
file_list
```

```python
file_list_csv = []
for file in file_list:
    if file.endswith('.csv'):
        file_list_csv.append(file)
file_list_csv
```

```python
file_list_csv[0]
```

```python
for i in range(1, len(file_list_csv)):
    print(i)
```

```python
first_df = pd.read_csv(file_list_csv[0])
```

```python
all_df = first_df
for i in range(1, len(file_list_csv)):
    temp = pd.read_csv(file_list_csv[i])
    all_df = pd.merge(all_df, temp, how='outer', on='글_내용')
mod_all_df = all_df.dropna()
mod_all_df
```

```python
title_df = mod_all_df['글_내용']
mod_all_df = mod_all_df.drop(['글_내용'], axis=1)
mod_all_df
```

```python
title_df
```

```python
mod_all_df.shape
```

```python
for i in range(mod_all_df.shape[0]):
    plt.figure(figsize=(12,8))
    plt.title(title_df[i])
    plt.plot(range(mod_all_df.shape[1]), mod_all_df.iloc[i, :])
    for j in range(mod_all_df.shape[1]):
        plt.text(i, mod_all_df.iloc[i,j], mod_all_df.iloc[i,j],                 # 좌표 (x축 = v, y축 = y[0]..y[1], 표시 = y[0]..y[1])
                fontsize = 9, 
                color='blue',
                horizontalalignment='center',  # horizontalalignment (left, center, right)
                verticalalignment='bottom')    # verticalalignment (top, center, bottom)
```
