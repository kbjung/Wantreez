```python
from selenium import webdriver
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
driver = webdriver.Chrome('chromedriver')
url = 'https://www.melon.com/chart/index.htm'
driver.get(url)
driver.maximize_window()
```

```python
soup = bs(driver.page_source, 'lxml')
soup.title.text
```

```python
# 날짜 수집
title = soup.find('span', 'yyyymmdd')
title.text.strip()
```

```python
# 시간 수집
time = soup.find('span', 'hhmm')
time.text.strip()
```

## columns 정보 수집

```python
# 순위
rank_col = soup.find('span', 'rank').text
rank_col
```

```python
# 곡정보
info_col = soup.find_all('div', 'wrap pd_l_12')
song_col = info_col[0].text
song_col
```

```python
# 앨범
album_col = info_col[1].text
album_col
```

```python
# 좋아요
like_col = soup.find('div', 'wrap pd_l_30').text
like_col
```

## 컬럼 명

```python
col_list = []
col_list.append(rank_col)
col_list.append(song_col)
col_list.append(album_col)
col_list.append(like_col)
col_list
```

```python
# 순위 정보
# 50위까지
top50_rank = soup.find_all('tr', 'lst50')
len(top50_rank)
```

```python
# 51~100위까지
top100_rank = soup.find_all('tr', 'lst100')
len(top100_rank)
```

```python
for one in top50_rank:
    temp = one.find('span', 'rank').text
    print(temp)
```

```python
for one in top100_rank:
    temp = one.find('span', 'rank').text
    print(temp)
```

## 순위 정보 수집

```python
rank_list = []
# 1~50위
for one in top50_rank:
    temp = one.find('span', 'rank').text
    rank_list.append(temp)
# 51~100위
for one in top100_rank:
    temp = one.find('span', 'rank').text
    rank_list.append(temp)

len(rank_list)    
```

```python
# 곡정보 수집
songs = soup.find_all('div', 'ellipsis rank01')
len(songs)
```

```python
for one in songs:
    temp = one.find('span').text
    print(temp)
```

## 곡 제목 수집

```python
song_list = []
songs = soup.find_all('div', 'ellipsis rank01')
for one in songs:
    temp = one.find('span').text.strip()
    song_list.append(temp)
len(song_list)
```

```python
# 가수 수집
artists = soup.find_all('div', 'ellipsis rank02')
len(artists)
```

```python
for one in artists:
    temp = one.find('a').text
    print(temp)
```

## 가수 수집

```python
artist_list = []
artists = soup.find_all('div', 'ellipsis rank02')
for one in artists:
    temp = one.find('a').text
    artist_list.append(temp)

len(artist_list)
```

```python
# 앨범 수집
albums = soup.find_all('div', 'ellipsis rank03')
len(albums)
```

```python
for one in albums:
    temp = one.find('a').text
    print(temp)
```

## 앨범 수집

```python
albums = soup.find_all('div', 'ellipsis rank03')
album_list = []
for one in albums:
    temp = one.find('a').text
    album_list.append(temp)

len(album_list)
```

```python
# 좋아요 수집
likes = soup.find_all('button', 'button_etc like')
len(likes)
```

## 좋아요 수집

```python
likes = soup.find_all('button', 'button_etc like')
like_list = []
for one in likes:
    temp = one.find('span', 'cnt').text.split('\n')[-1].replace(',','')
    temp = int(temp)
    like_list.append(temp)

len(like_list)
```

```python
dict = {col_list[0]:rank_list, col_list[1]:song_list ,'가수':artist_list, col_list[2]:album_list, col_list[3]:like_list}
df = pd.DataFrame(dict)
```
